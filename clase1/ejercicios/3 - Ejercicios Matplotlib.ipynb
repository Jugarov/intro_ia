Ejercicio 5:

Cree un gráfico en estructura de dos por dos usando **figure** para las siguientes funciones hiperbólicas en el dominio $ -1 \leq x \leq 1$

| Estructura del grafico | |
| ---------------------  | -- |
| $senh(x) \; vs \; x$ | $cosh(x) \; vs \; x$ |
| $tanh(x) \; vs \; x$ | $sech(x) \; vs \; x$ |

```python
x = np.linspace(-1, 1, 100)
sinh_array =  np.sinh(x)
cosh_array =  np.cosh(x)
tanh_array =  np.tanh(x)
sech_array =  1 / cosh_array

fig = plt.figure()

ax1 = fig.add_subplot(221)
ax1.plot(x, sinh_array)
ax1.set_xlabel('x')
ax1.set_ylabel('sinh(x)')
ax1.set_title('sinh')

ax2 = fig.add_subplot(222)
ax2.plot(x, cosh_array)
ax2.set_xlabel('x')
ax2.set_ylabel('cosnh(x)')
ax2.set_title('cosh')

ax3 = fig.add_subplot(223)
ax3.plot(x, tanh_array)
ax3.set_xlabel('x')
ax3.set_ylabel('tanh(x)')
ax3.set_title('tanh')

ax4 = fig.add_subplot(224)
ax4.plot(x, sech_array)
ax4.set_xlabel('x')
ax4.set_ylabel('sech(x)')
ax4.set_title('sech')
```
